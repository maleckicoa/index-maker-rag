```python
import db_funcs as db
import pickle
from df_funcs import *
from fmp_funcs import *
from urls import url_stock_list, url_ccy_info, url_hist_price, url_mcap_24, url_mcap_19_24, url_mcap_14_19, url_all_forex_pairs, url_forex, url_company_info, url_company_info_short
with open('all_symbols.pkl', 'rb') as file:
    all_symbols = pickle.load(file)
    
import asyncio
import aiohttp
import nest_asyncio
from datetime import datetime

# Allow asyncio to run in Jupyter Notebook
nest_asyncio.apply()

BASE_URL = "https://financialmodelingprep.com/api/v3/historical-price-full/"

async def fetch_historical_price(session, symbol, request_counter):
    """
    Fetches historical price data for a given stock symbol and updates the request counter.
    """
    url = f"{BASE_URL}{symbol}?from=2014-12-03&apikey={API_KEY}"
    print(url)
    async with session.get(url) as response:
        request_counter["count"] += 1  # Increment the request counter
        if response.status == 200:
            data = await response.json()
            print(f"[{datetime.now()}] Fetched historical data for {symbol}")
            return {symbol: data.get("historical", [])}
        else:
            print(f"[{datetime.now()}] Failed to fetch data for {symbol}: {response.status}")
            return {symbol: None}

async def fetch_historical_prices(symbols):
    """
    Fetches historical price data for a list of stock symbols, ensuring 5 requests per second and tracking the number of requests.
    """
    request_counter = {"count": 0}  # Shared counter for tracking requests
    results = []
    async with aiohttp.ClientSession() as session:
        for i in range(0, len(symbols), 15):  # Process 5 symbols at a time
            tasks = [fetch_historical_price(session, symbol, request_counter) for symbol in symbols[i:i+5]]
            batch_results = await asyncio.gather(*tasks)
            results.extend(batch_results)
            print(f"[{datetime.now()}] Requests made so far: {request_counter['count']}")
            await asyncio.sleep(0)  # Wait 1 second to stay within the 5 requests/second limit
    print(f"[{datetime.now()}] Total requests made: {request_counter['count']}")
    return results


results = await fetch_historical_prices(all_symbols)

# Print Final Results
for result in results:
    for symbol, historical_data in result.items():
        print(f"{symbol}: {len(historical_data)} historical records fetched.")

```
